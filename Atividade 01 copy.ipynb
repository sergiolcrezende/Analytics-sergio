8. Carregue as planilhas de 2016 e 2017 por nome em variáveis separadas. Calcule a média da coluna Age para as diferentes abas do arquivo.

```python
#ler planilhas
df_2016=pd.read_excel(path1,sheet_name='2016')
df_2017=pd.read_excel(path1,sheet_name='2017')

#calcular média
media_2016 = df_2016['Age'].mean()
media_2017 = df_2017['Age'].mean()

#imprimir média
print('Média 2016: {:.4f}'.format(media_2016),'e Média 2017: {:.4f}'.format(media_2017))
```

9. Concatene as diferentes abas do arquivo `fcc_survey.xlsx` em apenas um dataframe. Calcule a média da coluna Age para a concatenação das abas do arquivo.

```python
#ler planilhas
df_abas = pd.read_excel(path1,sheet_name=None)

#concatenar dataframe
df_concat= pd.concat(df_abas.values())

#calcular média
media = df_concat['Age'].mean()

#imprimir
print("Média da idade: {:.2f}".format(media))
```

10.  Utilizando uma query em SQL e o banco `sqlite-sakila.db` retorne as cidades e os países, mas apenas os países que começem com a letra **B**.

```python
pip install pysqlite3
```

```python
import sqlite3
```

```python
conn = sqlite3.connect(r'D:/instalar/sqllite/base/sqlite-sakila.db')
cursor = conn.cursor()
```

```python
sql = '''
         select country.country, city.city 
         from city 
         inner join country on city.country_id = country.country_id  
         where country.country like 'B%' 
         order by country.country
      '''

resultado= cursor.execute(sql)
for linha in resultado:
    print(linha)

conn.close()
```
